```python
# 1. 관련 데이터 수집
chembl_search_list = ['FTO', 'Alpha-ketoglutarate-dependent dioxygenase']
binding_db_folder = 'FTO'
target_protein = ['FTO', 'Alpha-ketoglutarate-dependent dioxygenase']

'''chembl_search_list = ['PDE','phosphodiesterase']
binding_db_folder = 'PDE'
target_protein = ['Phosphodiesterase 3B','PDE3B']'''

# 2. 데이터 수집 및 전처리
chembl_df, BDB_df = collect_data(chembl_search_list, binding_db_folder)
```

```python
# 3. 특이적 화합물 분석
chembl_df_unique_smile = unique_smiles_count(chembl_df, target_protein)
BDB_df_unique_smile = unique_smiles_count(BDB_df, target_protein)

chembl_fto_count = len(chembl_df)
bdb_fto_count = len(BDB_df)
total_data_points = chembl_fto_count + bdb_fto_count

print(f"\n=== 데이터 수 ===")
print(f"ChemBL: {chembl_fto_count}")
print(f"BindingDB: {bdb_fto_count}")
print(f"전체 데이터 포인트: {total_data_points}")

all_smiles = set()
all_smiles.update(chembl_df['canonical_smiles'].dropna())
all_smiles.update(BDB_df['canonical_smiles'].dropna())
unique_smiles_count = len(all_smiles)

print(f"Unique SMILES: {unique_smiles_count}")

if unique_smiles_count >= 500:
    print("→ Transfer learning 적용 가능")
elif unique_smiles_count >= 100:
    print("→ Transfer learning (소규모) 또는 Few-shot learning 고려")
else:
    print("→ Few-shot learning 적용 권장")
```
